```python
# store file in datafile variable using pandas
df = pd.read_csv('/kaggle/input/credit-risk-analysis-for-extending-bank-loans/bankloans.csv')
df.head() #first 5 rows
```

```python
#sum null values
df.isnull().sum()
```

```python
df.value_counts()
```

```python
# drop all 'unavailable' values in 'default' column since they are 1 or 0
df = df.dropna()
```

```python
# analyse relationship btn two cols using plot
fig,ax = plt.subplots(figsize = (20,10))
sns.lineplot(x='age', y='income', data = df,ax=ax)
```

```python
fig,ax = plt.subplots(figsize = (20,10))
sns.lineplot(x='age', y='debtinc', data = df,ax=ax)
```

```python
df['default'].value_counts()
```

```python
#create training and test data (80/20)
x = df.drop(['default'], axis = 1) #column
y = df['default']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42) #normalised random distribution
```

```python
#scale the parameters
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.fit_transform(xtest)
```

```python
# **ML model**
```

```python
Random Forest
```

```python
rfc = RandomForestClassifier(n_estimators = 200)
```

```python
rfc.fit(xtrain,ytrain)
```

```python
rfc.score(xtest,ytest)
```

```python
#refine parameters (may be overfit/underfit)
rfc2 = cross_val_score(estimator = rfc,X=xtrain, y=ytrain, cv=10) #10 fits
rfc2.mean()
```

```python
#enhance accuracy using Support Vector Machine
sv=SVC()
sv.fit(xtrain,ytrain)
sv.score(xtest,ytest)
```

```python
#SVM (hyper)parameters
model = GridSearchCV(sv,{
    'C': [0.1,0.2,0.4,0.8,1.2,1.8,4.0,7.0],
    'gamma': [0.1,0.4,0.8,1.0,2.0,3.0],
    'kernel': ['rbf','linear']
}, scoring = 'accuracy',cv=10)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.best_params_
```

```python
model2 = SVC(C = 0.1, gamma = 0.1, kernel = 'linear')
model2.fit(xtrain,ytrain)
model2.score(xtest,ytest)
```

```python
**Logistic Regression**
```

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
lr.score(xtest,ytest)
```

```python
yp = lr.predict(xtest)
c = confusion_matrix(ytest,yp)
fig, ax = plt.subplots(figsize = (20,10))
sns.heatmap(c,ax=ax)
```

```python
ypn = model2.predict(xtest)
cn= confusion_matrix(ytest,ypn)
fig, ax = plt.subplots(figsize = (20,10))
sns.heatmap(cn,ax=ax)
```
